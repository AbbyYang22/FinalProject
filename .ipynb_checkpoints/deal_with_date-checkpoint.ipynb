```python
# 先把長度小於6的都篩掉(轉格式or剔除)
j = 0
for i in range(len(data)) :   
    if len(data.construction_time[i].astype(str)) < 6 :
        if len(data.construction_time[i].astype(str)) == 4 :
            data.construction_time[i] = data.construction_time[i].astype(str)+'00'
            print(data.construction_time[i])
        elif data.construction_time[i].astype(str) == '90909' :
            data.construction_time[i] = '900909'
        else : 
            data.construction_time[i] = '1100100'
        j += 1
        
    else :
        continue
print(j)
```

```python
# 確認有無少於長度6的日期
j = 0
for i in range(len(data)) :   
    if len(data.construction_time[i].astype(str)) < 6 :
        j += 1
        print(data.construction_time[i])
        
    else :
        continue
print(j)
```

```python
# 確認trade_time有無小於長度6的日期
j = 0
for i in range(len(data)) :   
    if len(data.trade_time[i].astype(str)) < 6 :
        j += 1
        print(data.trade_time[i])
        
    else :
        continue
print(j)
```
