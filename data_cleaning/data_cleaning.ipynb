##　import cleaned_final_full_table

```python
clean = pd.read_csv('toTGOS/cleaned_final_full_table.csv')
clean.head()
```

```python
clean = clean.rename({"X": "id"}, axis = 1)
clean = clean.drop("Unnamed: 0", axis = 1)
```

```python
clean = clean.rename({"交易標的": "parking_space"}, axis = 1)
```

```python
# 0 -> 沒有車位
# 1 -> 有車位 
clean.parking_space.unique()
```

```python
clean.parking_space = clean.parking_space.map({2:1, 1:0})
```

```python
clean.parking_space.unique()
```

```python
# 0 -> 建物沒有隔間
# 1 -> 建物有隔間
clean['建物現況格局.隔間'].unique()
```

```python
clean['建物現況格局.隔間'] = clean['建物現況格局.隔間'].map({"有": 1, "0": 0})
```

```python
clean['建物現況格局.隔間'].unique()
```

```python
clean.loc[:, '建物型態'].unique() # clean['建物型態'].unique()
```

```python
# 0 -> 住宅大樓(11層含以上有電梯)
# 1 -> 公寓(5樓含以下無電梯)
# 2 -> 華廈(10層含以下有電梯)
# 3 -> 套房(1房1廳1衛)
# 4 -> 透天厝
# 5 -> 其他
clean['建物型態'] = clean['建物型態'].map(
    {"住宅大樓(11層含以上有電梯)": 0, "公寓(5樓含以下無電梯)": 1, "華廈(10層含以下有電梯)": 2, "套房(1房1廳1衛)": 3, "透天厝": 4, "其他": 5}
)
```

```python
clean['建物型態'].value_counts()
```

```python
clean['單價元平方公尺'] = clean['單價元平方公尺'] * 3.3058
```

```python
clean = clean.rename({"單價元平方公尺": "每坪價格"}, axis = 1)
```

```python
clean.columns
```

```python
clean['主要用途'].unique()
```

```python
sum((clean['主要用途'] == '住家用'))
```

```python
sum((clean['主要用途'] == '社區住宅'))
```

```python
sum((clean['主要用途'] == '住宅'))
```

```python
sum((clean['主要用途'] == '國民住宅'))
```

```python
sum((clean['主要用途'] == '住宅，集合住宅'))
```

```python
clean_1 = clean[clean['主要用途'] == '住家用']
clean_2 = clean[clean['主要用途'] == '社區住宅']
clean_3 = clean[clean['主要用途'] == '住宅']
clean_4 = clean[clean['主要用途'] == '國民住宅']
clean_5 = clean[clean['主要用途'] == '住宅，集合住宅']
```

```python
# 合併主要用途dataframe
houstList = [clean_1, clean_2, clean_3, clean_4, clean_5]
clean_final = pd.concat(houstList)
```

```python
# drop預設把index新增成原df的column
clean_final = clean_final.reset_index(drop = True)
```

```python
clean_final
```

```python
# clean_final[clean_final['建物現況格局.廳'] > 10].index，取得符合條件的index
clean_final = clean_final.drop(index = (clean_final[clean_final['建物現況格局.廳'] > 10].index))
clean_final = clean_final.drop(index = (clean_final[clean_final['建物現況格局.衛'] > 10].index))
clean_final = clean_final.drop(index = (clean_final[clean_final['建物現況格局.房'] > 10].index))
```

```python
clean_final
```

```python
clean_final.to_csv('cleaned_final_full_table.csv', index = False, encoding = 'UTF-8')
```
