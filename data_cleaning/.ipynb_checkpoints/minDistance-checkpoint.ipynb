```python
def getDistance(latA,lonA,latB,lonB):

    ra = 6378140 # 赤道半徑

    rb = 6356755 # 極   半徑

    flatten = (ra - rb) / ra # Partial rate of the earth

  # change angle to radians

    radLatA = math.radians(latA)

    radLonA = math.radians(lonA)

    radLatB = math.radians(latB)

    radLonB = math.radians(lonB)

    pA = math.atan(rb / ra * math.tan(radLatA))

    pB = math.atan(rb / ra * math.tan(radLatB))

    x = math.acos(math.sin(pA) * math.sin(pB) + math.cos(pA) * math.cos(pB) * math.cos(radLonA - radLonB))

    c1 = (math.sin(x) - x) * (math.sin(pA) + math.sin(pB)) ** 2 / math.cos(x / 2) ** 2

    c2 = (math.sin(x) + x) * (math.sin(pA) - math.sin(pB)) ** 2 / math.sin(x / 2) ** 2

    dr = flatten / 8 * (c1 - c2)

    distance = ra * (x + dr)

    distance = round(distance / 1000,4)

    return distance
```

```python
time1 = time.time()
try : 
    for i in range(len(df)) :
        distanceList_hospital = []
        distanceList_gas = []
        distanceList_inci = []
        distanceList_mrt = []
        distanceList_park = []
        distanceList_mart = []
        distanceList_school = []
        for j in range(len(hospital)) : 
            latAddress = df.iloc[i, 4]
            lonAddress = df.iloc[i, 3]
            latTarget = hospital.iloc[j, 4]
            lonTarget = hospital.iloc[j, 3]
            distanceList_hospital.append(getDistance(latAddress, lonAddress, latTarget, lonTarget))
        df.iloc[i, 6] = min(distanceList_hospital)
        for j in range(len(gas)) : 
            latAddress = df.iloc[i, 4]
            lonAddress = df.iloc[i, 3]
            latTarget = gas.iloc[j, 1]
            lonTarget = gas.iloc[j, 2]
            distanceList_gas.append(getDistance(latAddress, lonAddress, latTarget, lonTarget))
        df.iloc[i, 9] = min(distanceList_gas)
        for j in range(len(inci)) : 
            latAddress = df.iloc[i, 4]
            lonAddress = df.iloc[i, 3]
            latTarget = inci.iloc[j, 1]
            lonTarget = inci.iloc[j, 2]
            distanceList_inci.append(getDistance(latAddress, lonAddress, latTarget, lonTarget))
        df.iloc[i, 5] = min(distanceList_inci)
        for j in range(len(mrt)) : 
            latAddress = df.iloc[i, 4]
            lonAddress = df.iloc[i, 3]
            latTarget = mrt.iloc[j, 1]
            lonTarget = mrt.iloc[j, 2]
            distanceList_mrt.append(getDistance(latAddress, lonAddress, latTarget, lonTarget))
        df.iloc[i, 7] = min(distanceList_mrt)
        for j in range(len(park)) : 
            latAddress = df.iloc[i, 4]
            lonAddress = df.iloc[i, 3]
            latTarget = park.iloc[j, 1]
            lonTarget = park.iloc[j, 2]
            distanceList_park.append(getDistance(latAddress, lonAddress, latTarget, lonTarget))
        df.iloc[i, 10] = min(distanceList_park)
        for j in range(len(school)) : 
            latAddress = df.iloc[i, 4]
            lonAddress = df.iloc[i, 3]
            latTarget = school.iloc[j, 1]
            lonTarget = school.iloc[j, 2]
            distanceList_school.append(getDistance(latAddress, lonAddress, latTarget, lonTarget))
        df.iloc[i, 11] = min(distanceList_school)
         for j in range(len(mart)) : 
            latAddress = df.iloc[i, 4]
            lonAddress = df.iloc[i, 3]
            latTarget = mart.iloc[j, 1]
            lonTarget = mart.iloc[j, 2]
            distanceList_mart.append(getDistance(latAddress, lonAddress, latTarget, lonTarget))
        df.iloc[i, 8] = min(distanceList_mart)
except ZeroDivisionError :
    print(f"第{i}列有問題")
    print(f"總共花費{time.time()-time1}")
```
