# filter general_model_section

```python
general_model_section = data.drop(columns = {'floor', 'yearMonth'})
```

```python
general_model_section.columns
```

```python
groupby_section = general_model_section.groupby('section')
```

```python
groupby_section.size()
```

```python
for i in range(1,29) :
    if i not in (11,17) :
        section = groupby_section.get_group(i)
        section = section.drop('section', axis = 1)
        section.to_csv(f'model_data/general_model_section/general_model_section_{i}.csv', index = False, encoding = 'utf-8')
    else :
        continue
```

# filter general_model_type

```python
general_model_type = data.drop('yearMonth', axis = 1)
```

```python
general_model_type
```

```python
groupby_type = general_model_type.groupby('type')
```

```python
groupby_type.size()
```

```python
for i in range(0, 5):
    types = groupby_type.get_group(i)
    types = types.drop('type', axis = 1)
    types.to_csv(f'model_data/general_model_type/general_model_type_{i}.csv', index = False, encoding = 'utf-8')
```
