# Add column yearMonth for format xxx-xx/xx-xx

```python
data.trade_time = data.trade_time.astype(str)
```

```python
data = data.reset_index(drop = True)
```

```python
from time import time
time1 = time()
for i in range(len(data)) :
    if len(data.trade_time[i]) == 7:
        data.yearMonth[i] = data.trade_time[i][0:3]+'-'+data.trade_time[i][3:5]
    else :
        data.yearMonth[i] = data.trade_time[i][0:2]+'-'+data.trade_time[i][2:4]
        
print(f"總共花費{time()-time1}")
```
