```python
# section forcast plot
for i in range(29):
    if i == 28:
        
        print(i)
        areaName = nameList[i-1]
        
        path = 'C:/Users/Peter/Desktop/final_project/Data/time_model_section/time_model_section_' + str(i) + '.csv'
        i = pd.read_csv(path)
        
        # select required columns and rename to the required names
        i = i[['yearMonth', 'unit_price_ping']]
        i.columns = ['ds', 'y']
        
        # change the date format
        year = i.ds.str.split("-", expand=True)[0]
        month = i.ds.str.split("-", expand=True)[1]
        
        year = year.astype(int) + 1911
        i['ds'] = year.astype(str) + '-' + month.astype(str)
        
        # transfer into date format: year-month-day
        i['ds'] = pd.to_datetime(i['ds'])
        
        # mean value & groupby date
        y = i.groupby('ds').mean()[['y']]
        y['ds'] = y.index
        y = y.reset_index(drop=True)
        y = y[['ds','y']]
        
        # time series
        prophet = Prophet()
        prophet.fit(y)
        future = prophet.make_future_dataframe(periods=36, freq='MS') # for the next three years
        forecast = prophet.predict(future)
        forecast[['ds','yhat','yhat_lower','yhat_upper']].head()
        
        # plot
        
        fig = prophet.plot(forecast)
        ax = fig.gca()
        ax.set_title(areaName, size=25, fontproperties=ChineseFont1)
        ax.set_xlabel("Year", size=12)
        ax.set_ylabel("Price/Ping", size=12)
        ax.tick_params(axis="x", labelsize=8)
        ax.tick_params(axis="y", labelsize=8)
#         fig.set_size_inches(12, 10)
        fig.savefig(areaName, dpi=200, bbox_inches='tight')
```
